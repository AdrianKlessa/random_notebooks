```python
import matplotlib.pyplot as plt
print("test")
```

```python
def linear_congruence():
    m = (2**16)-1
    a = 75
    c =74
    x = 2
    while True:
        x=(a*x+c)%m
        yield x
```

```python
numbers_list = []
g = linear_congruence()
for x in range (10000):
    numbers_list.append(next(g))
```

```python
plt.title("Numbers generated over time")
plt.plot(numbers_list)
plt.show()
```

```python
plt.title("Distribution of numbers")
plt.hist(numbers_list)
plt.show()
```
