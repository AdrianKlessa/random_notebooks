## Linear Congruential Generator

Linear congruences is a method of generating pseudo-random numbers using a simple recurrence relation.
<br>
Obviously not suitable for cryptographic purposes.
<br>
Mersenne twister is a popular alternative.

```python
import matplotlib.pyplot as plt
```

```python
def linear_congruence(m,a,c,x):
    while True:
        x=(a*x+c)%m
        yield x
```

```python
numbers_list = []
g = linear_congruence((2**16)-1,75,74,2) #ZX81
for x in range (10000):
    numbers_list.append(next(g))
```

```python
print("First 100 generated numbers:")
print(numbers_list[:100])
```

```python
plt.title("Numbers generated over time with ZX81 settings")
plt.plot(numbers_list)
plt.show()
```

```python
plt.title("Distribution of numbers with ZX81 settings")
plt.hist(numbers_list, bins=50)
plt.show()
```

```python
numbers_list = []
g = linear_congruence((2**31)-1,48271,0,2) #C++11's minstd_rand
for x in range (10000):
    numbers_list.append(next(g))
```

```python
plt.title("Numbers generated over time with C11's settings")
plt.plot(numbers_list)
plt.show()
```

```python
plt.title("Distribution of numbers with C11's settings")
plt.hist(numbers_list, bins=50)
plt.show()
```
