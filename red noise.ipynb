Red noise, unlike white noise is correlated in time with its previous value. 
A time series may look like it has a certain trend, even though it actually is just random noise.

```python
import matplotlib.pyplot as plt
import random
```

White noise time series can be generated by picking samples from the gaussian distribution at each step

# White noise time series example:

```python
x_values = [0]

for i in range(100):
    x_values.append(random.uniform(-1,1))

plt.plot(x_values)
```

Red noise time series can be generated by getting a starting value, and then:
1. Sampling a random value from the gaussian distribution
2. Adding this sampled value to the previous value in the time series. This sum is the new value.

# Red noise time series examples:

```python
x_values = [0]

for i in range(100):
    x_values.append(random.uniform(-1,1)+x_values[-1])

plt.plot(x_values)
```

```python
x_values = [0]

for i in range(1000):
    x_values.append(random.uniform(-1,1)+x_values[-1])

plt.plot(x_values)
```

## As can be seen below, over large periods of time this noise tends to "wander off" from zero:

```python
x_values = [0]

for i in range(100000):
    x_values.append(random.uniform(-1,1)+x_values[-1])

plt.plot(x_values)
```
